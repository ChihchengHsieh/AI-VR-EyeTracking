## Making the stay_id link between the 0.4 and 1.0 versions.

```python
import pandas as pd
import time
import numpy as np
import os
import random
from datetime import datetime
import math
from tqdm.notebook  import tqdm
```

```python
# Define the path for all of the folders.
ED_FOLDER_PATH = "E:/AI-VR dataset/MIMIC-IV ED"
CLINICAL_FOLDER_PATH = "E:/AI-VR dataset/MIMIC-IV Clinical Database"
CXR_FOLDER_PATH = "E:/AI-VR dataset/MIMIC-CXR-JPG/physionet.org/files/mimic-cxr-jpg/2.0.0"
EYEGAZE_FOLDER_PATH = "E:/AI-VR dataset/eye-gaze-data-for-chest-x-rays-1.0.0"
REFLACX_FOLDER_PATH = "E:/AI-VR dataset/reflacx-reports-and-eye-tracking-data-for-localization-of-abnormalities-in-chest-x-rays-1.0.0/reflacx-reports-and-eye-tracking-data-for-localization-of-abnormalities-in-chest-x-rays-1.0.0"
XAMI_MIMIC_PATH = "E:/AI-VR dataset/XAMI-MIMIC"
```

```python
transfers_df_0dot4 = pd.read_csv(f"transfers-0.4.csv.gz", compression='gzip', header=0, sep=',', quotechar='"')
```

```python
ED_edstays_df = pd.read_csv(f"{ED_FOLDER_PATH}/edstays.csv")
```

```python
new_to_older_stay_id = {}
```

```python
len(ED_edstays_df)
```

```python
# The other way around.
transfer_ed_df = transfers_df_0dot4[transfers_df_0dot4["eventtype"] == "ED"]


with tqdm(total= len(ED_edstays_df)) as pbar:
    for ed_stay_instance in ED_edstays_df.iloc:
        subject_id = ed_stay_instance['subject_id']

        matches_transfered = transfer_ed_df[
            (transfer_ed_df['subject_id'] == subject_id)
        ]

        if len(matches_transfered) > 0:
            matches_transfered = matches_transfered[(matches_transfered['intime'] == ed_stay_instance['intime']) & (matches_transfered['outtime'] == ed_stay_instance['outtime'])]
            
        if (len(matches_transfered) > 0) and (not (matches_transfered.iloc[0]['transfer_id'] in new_to_older_stay_id.values())):
            new_to_older_stay_id[ed_stay_instance['stay_id']] = matches_transfered.iloc[0]['transfer_id']

        pbar.update(1)
```

```python
save_new_to_older_stay_id ={}
```

```python
for k, v in new_to_older_stay_id.items():
    save_new_to_older_stay_id[int(k)] = int(v)
```

```python
import json

with open('new_to_older_stay_id.json', 'w') as fp:
    json.dump(save_new_to_older_stay_id, fp)
```

```python
with open('new_to_older_stay_id.json', 'r') as fp:
    new_to_older_stay_id_load_test = json.load(fp)
```
